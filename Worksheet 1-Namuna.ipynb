Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)

T = BTAT
.

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np

# Given matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I (Identity Matrix)
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print("A * A^-1 = Identity Matrix:")
print(identity_matrix)

# 2. Prove AB != BA (Check if AB equals BA)
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:")
print(AB)
print("\nBA:")
print(BA)

# Check if AB == BA
if np.array_equal(AB, BA):
    print("\nAB equals BA")
else:
    print("\nAB does not equal BA")

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("\n(AB)^T:")
print(AB_T)
print("\nB^T * A^T:")
print(BT_AT)

# Check if (AB)^T == B^T * A^T
if np.array_equal(AB_T, BT_AT):
    print("\n(AB)^T equals B^T * A^T")
else:
    print("\n(AB)^T does not equal B^T * A^T")

# 4. Solve the system of equations using Inverse Method (AX = B)
# System of equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent the system in matrix form: AX = B
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Solve for X = A^-1 * B
X = np.dot(np.linalg.inv(A_eq), B_eq)

print("\nSolution of the system of equations (X = [x, y, z]):")
print(X)
```

4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import numpy as np
import time

# 1. Element-wise Addition

# Using Python lists
list1 = [i for i in range(1000000)]
list2 = [i for i in range(1000000)]

start_time = time.time()
list_sum = [list1[i] + list2[i] for i in range(1000000)]
end_time = time.time()
python_list_addition_time = end_time - start_time
print(f"Python list element-wise addition time: {python_list_addition_time:.5f} seconds")

# Using NumPy arrays
array1 = np.arange(1000000)
array2 = np.arange(1000000)

start_time = time.time()
numpy_sum = array1 + array2
end_time = time.time()
numpy_array_addition_time = end_time - start_time
print(f"NumPy array element-wise addition time: {numpy_array_addition_time:.5f} seconds")

# 2. Element-wise Multiplication

# Using Python lists
start_time = time.time()
list_multiply = [list1[i] * list2[i] for i in range(1000000)]
end_time = time.time()
python_list_multiplication_time = end_time - start_time
print(f"Python list element-wise multiplication time: {python_list_multiplication_time:.5f} seconds")

# Using NumPy arrays
start_time = time.time()
numpy_multiply = array1 * array2
end_time = time.time()
numpy_array_multiplication_time = end_time - start_time
print(f"NumPy array element-wise multiplication time: {numpy_array_multiplication_time:.5f} seconds")

# 3. Dot Product

# Using Python lists
start_time = time.time()
list_dot_product = sum([list1[i] * list2[i] for i in range(1000000)])
end_time = time.time()
python_list_dot_product_time = end_time - start_time
print(f"Python list dot product time: {python_list_dot_product_time:.5f} seconds")

# Using NumPy arrays
start_time = time.time()
numpy_dot_product = np.dot(array1, array2)
end_time = time.time()
numpy_array_dot_product_time = end_time - start_time
print(f"NumPy array dot product time: {numpy_array_dot_product_time:.5f} seconds")

# 4. Matrix Multiplication

# Create two 1000x1000 matrices using Python lists
matrix1 = [[i for i in range(1000)] for j in range(1000)]
matrix2 = [[i for i in range(1000)] for j in range(1000)]

# Using Python lists for matrix multiplication
start_time = time.time()
matrix_product = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(1000)) for j in range(1000)] for i in range(1000)]
end_time = time.time()
python_matrix_multiplication_time = end_time - start_time
print(f"Python list matrix multiplication time: {python_matrix_multiplication_time:.5f} seconds")

# Using NumPy arrays for matrix multiplication
matrix1_np = np.arange(1000000).reshape(1000, 1000)
matrix2_np = np.arange(1000000).reshape(1000, 1000)

start_time = time.time()
numpy_matrix_product = np.dot(matrix1_np, matrix2_np)
end_time = time.time()
numpy_matrix_multiplication_time = end_time - start_time
print(f"NumPy array matrix multiplication time: {numpy_matrix_multiplication_time:.5f} seconds")
```
